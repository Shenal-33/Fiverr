```python
#A-5
# Note: Sometimes the colors do not load up for some reason, try running the cell
corr = df.corr(method = 'pearson')
display(corr.style.background_gradient(cmap="coolwarm").set_precision(4))

corr2 = corr.copy()
corr2['corAbs'] = corr2['Final'].apply(lambda x : abs(x))
corr2.sort_values(by='corAbs', inplace=True, ascending=False)   

topThree = list(corr2.iloc[1:4].index)
print('The most correlated colmuns with Final are:')
for (i,col) in enumerate(topThree) :
    print(f'{i+1}. {col}')
    
corr = corr.drop(['Quiz-2','Quiz-3'], axis = 1)
corr = corr.drop(['Quiz-2','Quiz-3'], axis = 0)
display(corr.style.background_gradient(cmap="coolwarm").set_precision(4))
print("\nAll pairs above are strong correlated except when there are Quiz-2 and Quiz-3" )    
print('\n\nThe strongest pair of scores are : \nFinal and Major-2')
```

```python
#A-6
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

num_columns = df.select_dtypes(exclude='object').columns
pca_input_columns=num_columns.drop(['Final'])

pca = PCA(n_components=2)

principalComponents = pca.fit_transform(df[pca_input_columns]) 

df['PC1'] = principalComponents[:,0]
df['PC2'] = principalComponents[:,1]
display(df)
df["Final"] = Finals
df['finalDef'] = df['Final'].apply(lambda x : 'final>85' if x > 85 else '84>final>65' if x >65 else 'final<65')
sns.relplot(x='PC1', y='PC2',palette=['g','r', 'b'], hue='finalDef',data=df)
sns.relplot(x='PC1', y='PC2',palette=['g','r', 'b'], hue='finalDef',col='Gender',data=df)
sns.relplot(x='PC1', y='PC2',palette=['g','r', 'b'], hue='finalDef',col='Location',data=df)
plt.show()

df.drop(["finalDef"],axis = 1, inplace = True)

for ind,var in enumerate(pca.explained_variance_ratio_):
    print(f'The variance captured by pca {ind+1} is: {var*100: 0.2f}%')

pca_1_coefficients=np.round(pca.components_[0],4)
zipList=list(zip(pca_input_columns,pca_1_coefficients))
print(zipList)
```

```python
#B-1
import pandas as pd
df1 = pd.read_csv("HW5DataB.csv", delimiter = ',')
display(df1.head())
df1.info()
print(f"The number of rows are {len(df1.index)}, and the number of columns are {len(df1.columns)}")
## there is no any missing value
display(df1.describe().T)
```

```python
#B-2
display(df1.head())
## Quiz-1   type Numrical
## Quiz-2   type Numrical
## Quiz-3   type Numrical
## Quiz-4   type Numrical
## Major-1  type Numrical
## Major-2  type Numrical
## Major-3  type Numrical
## Final    type Numrical
df1.info()
## there is no inconsistency
```

```python
#B-3
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
df1=pd.DataFrame(scaler.fit_transform(df1),columns=df1.columns)
display(df1.describe().T)
display(df1)
```

```python
#B-4
df2 = pd.read_csv('HW5DataC.csv')
df2=pd.DataFrame(scaler.transform(df2),columns=df2.columns)
display(df2.describe())
display(df2)
```

```python
#B-5

Xo = df1.iloc[:,0:-1].values
y = df1.iloc[:,-1].values
X = np.c_[np.ones(len(y)), Xo]

best_beta = np.linalg.inv(X.T @ X) @ X.T @ y

print('The closed form estimates are:', np.round(best_beta,2).tolist()) 
```

```python
#B-6
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error

reg = LinearRegression().fit(Xo, y)

best_beta =  np.round(reg.coef_,2)
best_beta_0 = np.round(reg.intercept_,2)



Xo2 = df2.iloc[:,0:-1].values
y2 = df2.iloc[:,-1].values

reg1 = LinearRegression(fit_intercept=False).fit(Xo, y)
y_pred1 = reg1.predict(Xo2)
print(f'The best values for the estimates are:   ', best_beta_0, best_beta.tolist())
print('The MSE using OLS is:                    ', mean_squared_error(y2,y_pred1))
## there is no diffrence between the data in B-5 or B-6
```

```python
#B-7
from sklearn.linear_model import RidgeCV
reg2 = RidgeCV(alphas=[1e-3, 1e-2, 1e-1, 1e0, 1e1, 1e2, 1e3], fit_intercept=False,cv=10).fit(Xo, y)
y_pred2 = reg2.predict(Xo2)

best_beta2 =  np.round(reg2.coef_,2)
best_beta_02 = np.round(reg2.intercept_,2)

print(f'The best values for the estimates are:   ', best_beta_02, best_beta2.tolist())
print('The MSE using Ridge is:                  ', mean_squared_error(y2, y_pred2))
print('The best penalty coefficient is equal to:', reg2.alpha_)
```

```python
#B-8
from sklearn.linear_model import LassoCV
reg3 = LassoCV(alphas=[1e-3, 1e-2, 1e-1, 1e0, 1e1, 1e2, 1e3], fit_intercept=False,cv=10, random_state=0).fit(Xo, y)
y_pred3 = reg3.predict(Xo2)

best_beta3 =  np.round(reg3.coef_,2)
best_beta_03 = np.round(reg3.intercept_,2)

print(f'The best values for the estimates are:   ', best_beta_03, best_beta3.tolist())
print('The MSE using Lasso is:                  ', mean_squared_error(y2, y_pred3))
print('The best penalty coefficient is equal to:', reg3.alpha_)
```

```python
#B-9
print('The MSE using OLS is:  ', mean_squared_error(y2, y_pred1))
print('The MSE using Ridge is:', mean_squared_error(y2, y_pred2))
print('The MSE using Lasso is:', mean_squared_error(y2, y_pred3))
print("\nThe lowest MSE is when we use Ridge")
print("The values of MSE using OLS and Ridge are very close to each other")
print('\nThe of OLS Regression coefficient estimates are:   \n', reg1.coef_)
print('\nThe of Ridge Regression coefficient estimates are: \n', reg2.coef_)
print('\nThe of Lasso Regression coefficient estimates are: \n', reg3.coef_)
```
