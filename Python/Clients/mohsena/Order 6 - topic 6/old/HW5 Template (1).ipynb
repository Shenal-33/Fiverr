# Problem #A

```python
#A-1
import pandas as pd
df  = pd.read_csv('HW5DataA.csv')
display(df.head())
print(f'The number of colmuns is {len(df.columns)}')
print(f'The number of rows is {len(df.index)}')
display(df.info())
print('No null values')

print('Desciption for the numeric colmuns: ')
display(df.describe(include='number'))
print('Desciption for the non-numeric colmuns: ')
display(df.describe(include='object'))

```

```python

#A-2
colType = {
    'Gender' : 'Cat',
    'Location' : 'Cat',
    'Quiz-1' : 'num',
    'Quiz-2' : 'num',
    'Quiz-3' : 'num',
    'Quiz-4' : 'num',
    'Major-1' : 'num',
    'Major-2' : 'num',
    'Major-3' : 'num',
    'Final' : 'num',
}

numCol = df.select_dtypes(include='number').columns
catCol = df.select_dtypes(include='object').columns


for col in catCol:
    if(colType[col] != 'Cat'):
        print(f'column {col} is NOT identified correctly')
    else:
        print(f'column {col} is identified correctly')



for col in numCol:
    if(colType[col] != 'num'):
        print(f'column {col} is NOT identified correctly')
    else:
        print(f'column {col} is identified correctly')
        



```

```python
#A-3

from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()


for col in numCol :
    scaler = StandardScaler()
    scaler.fit(df[[col]])
    df[col] = scaler.transform(df[[col]])

    
display(df)
```

```python
#A-4

import seaborn as sn
import matplotlib.pyplot as plt

figure = plt.figure()
sn.pairplot(data=df, hue='Gender', x_vars=numCol, kind='scatter', diag_kind='kde')
```

```python
#A-5

corelations = df[numCol].corr()
display(corelations)

corelations['corAbs'] = corelations['Final'].apply(lambda x : abs(x))

corelations.sort_values(by='corAbs', inplace=True, ascending=False)


topThree = list(corelations.iloc[1:4].index)
print('The most correlated colmuns are:')
for (i,col) in enumerate(topThree) :
    print(f'{i+1}. {col}')
    
    
print('\n\n\nPait of scores strongly correlated are : \nMajor1 and Major2 \nMajor1 and Major3')
```

```python
#A-6
from sklearn.decomposition import PCA
import numpy as np
pca = PCA(n_components=2)

targetCol = numCol.drop(['Final'])
firstTwoPcas = pca.fit_transform(df[targetCol])

df['PCA1'] = firstTwoPcas[:, 0]
df['PCA2'] = firstTwoPcas[:, 1]

figure = plt.figure()


df['finalDef'] = df['Final'].apply(lambda x : 'final>85' if x > 85 else '84>final>65' if x >65 else 'final<65')
sn.scatterplot(x= 'PCA1', y='Final', data=df, palette=['g','r', 'b'], hue='finalDef')
figure.show()
figure = plt.figure()

sn.scatterplot(x= 'PCA2', y='Final', data=df,  palette=['g','r', 'b'], hue='finalDef')
figure.show()


sn.relplot(x= 'PCA1', y='Final', data=df, palette=['g','r', 'b'], hue='finalDef', col='Gender')


sn.relplot(x= 'PCA1', y='Final', data=df, palette=['g','r', 'b'], hue='finalDef', col='Location')


sn.relplot(x= 'PCA2', y='Final', data=df, palette=['g','r', 'b'], hue='finalDef', col='Gender')


sn.relplot(x= 'PCA2', y='Final', data=df, palette=['g','r', 'b'], hue='finalDef', col='Location')
figure.show()



pca_1_coefficients=np.round(pca.components_[0],4)

zipList=list(zip(targetCol,pca_1_coefficients))
ndf=pd.DataFrame.from_records(zipList,columns=['field','coeff'])

display(ndf.sort_values(by=['coeff'], ascending=False))  # remember to consider abs() for negative values...
for ind,var in enumerate(pca.explained_variance_ratio_):
    print(f'The variance captured by pca {ind+1} is: {var*100: 0.2f}%')
```

```python
#B-4

scaler = StandardScaler()
scaler.fit(df2)
 
    
df2[df2.columns] = scaler.transform(df2[df2.columns])

display(df2)
display(df2.describe())
```

```python
#B-5
Xo = df2.iloc[:,0:-1].values
y = df2.iloc[:,-1].values
X = np.c_[np.ones(len(y)), Xo]



best_beta = np.linalg.inv(X.T @ X) @ X.T @ y

print('The closed form estimates are:', np.round(best_beta,2).tolist()) 
```

```python
#B-6
from sklearn.linear_model  import LinearRegression
from sklearn.metrics import mean_squared_error
ols = LinearRegression()
ols.fit(Xo, y)

best_beta =  np.round(ols.coef_,2)
best_beta_0 = np.round(ols.intercept_,2)
mean_squared_error()
print(f'The best values for the estimates are :', best_beta_0, best_beta.tolist())
```
